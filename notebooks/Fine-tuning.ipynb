## Fine Tune the Named Entity Recognition (NER) Model (Task 3)

Transformers in fine-tuning refer to the use of pre-trained transformer models, which are advanced neural network architectures designed for processing sequential data, particularly in natural language processing tasks. These models, such as BERT, GPT, or T5, are initially trained on large datasets to learn general language understanding. 

Fine-tuning involves adapting these pre-trained models to specific tasks or domains by further training them on smaller, task-specific datasets. This process leverages the model's pre-existing knowledge, allowing for improved performance on specialized tasks with less data and computational resources than training from scratch. 

Fine-tuning transformers typically involves adjusting the model's parameters or adding task-specific layers to tailor the model's capabilities to tasks like text classification, named entity recognition, or question answering, among others.

```python
# load_dataset function to load the Yelp review dataset of hugging face pre-trained models
from datasets import load_dataset

dataset = load_dataset("yelp_review_full")
dataset["train"][100]
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")


def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)


tokenized_datasets = dataset.map(tokenize_function, batched=True)
```

```python
small_train_dataset = tokenized_datasets["train"].shuffle(seed=42).select(range(1000))
small_eval_dataset = tokenized_datasets["test"].shuffle(seed=42).select(range(1000))
```

#### Train the model /dataset

```python
pip install torch torchvision torchaudio --index-url https://download.pytorch.org/whl/cu118
```

```python
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained("bert-base-cased", num_labels=5)
```

### Training hyperparameters

```python
from transformers import TrainingArguments

training_args = TrainingArguments(output_dir="test_trainer")
```

### Metrics

```python
import numpy as np
import evaluate

metric = evaluate.load("accuracy")
```

```python
def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)
    return metric.compute(predictions=predictions, references=labels)
```

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(output_dir="test_trainer", evaluation_strategy="epoch")
```

### Trainer

```python
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=small_train_dataset,
    eval_dataset=small_eval_dataset,
    compute_metrics=compute_metrics,
)
```

```python
trainer.train()
```

```python
eval_results = trainer.evaluate()
print(f"Evaluation results: {eval_results}")
```

```python
# Save the fine-tuned model and tokenizer
trainer.save_model('./fine_tuned_model')
tokenizer.save_pretrained('./fine_tuned_model')
```
